```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_pokemon.csv')
df.head(2)
```

```python
for i in df['Type 1'].unique():
    print(i)
```

```python
values = [df.loc[df['Type 1'] == i].count()[1] for i in df['Type 1'].unique()]  # Parts of a pie

labels = [i for i in df['Type 1'].unique()]  # Labels of parts

autopct = '%.1f %%'  # Show percentage value

pctdistance = 0.7  # Distance from center

explode = [0.05 for i in df['Type 1'].unique()]  # Explode a part
```

```python
plt.figure(figsize=(14, 9))

plt.pie(values, labels = labels, autopct=autopct, pctdistance=pctdistance, explode=explode)

# Making a donut by drawing a white circle in the middle of the figure
centre_circle = plt.Circle((0,0), 0.50, color='white', fc='white', linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.show()
```
